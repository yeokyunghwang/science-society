```python
DATA_ROOT = "./data"
YEARS = range(1990, 1996)

PKL_TPL = {
    "news":  "news_network_{year}_m0.0-M1.0.pkl",
    "paper": "paper_network_{year}_m0.0-M0.5.pkl",
}

rows = []

for domain, tpl in PKL_TPL.items():
    for year in YEARS:
        pkl_path = os.path.join(
            DATA_ROOT,
            domain,
            tpl.format(year=year)
        )

        if not os.path.exists(pkl_path):
            print(f"[SKIP] missing {pkl_path}")
            continue

        obj = pd.read_pickle(pkl_path)
        G = obj["G"]

        rows.append({
            "domain": domain,
            "year": year,
            "num_nodes": G.number_of_nodes(),
            "num_edges": G.number_of_edges(),
        })

df = pd.DataFrame(rows).sort_values(["domain", "year"])
df
```

```python
import pandas as pd

def build_yearly_table(
    target: str,
    years,
    base_dir="./outputs/mask_pred/dist=10",
    top_k=20,
):
    """
    returns: news_df, paper_df
    rows = rank
    cols = (year_node, year_prob)
    """
    news_frames = []
    paper_frames = []

    for y in years:
        path = f"{base_dir}/year={y}/target={target}.tsv"
        df = pd.read_csv(path, sep="\t").head(top_k)

        # rank를 index로 통일
        df = df.set_index("rank")

        # --- NEWS ---
        news_part = df[["news_neighbor", "news_avg_prob"]].copy()
        news_part.columns = [f"{y}_node", f"{y}_prob"]
        news_frames.append(news_part)

        # --- PAPER ---
        paper_part = df[["paper_neighbor", "paper_avg_prob"]].copy()
        paper_part.columns = [f"{y}_node", f"{y}_prob"]
        paper_frames.append(paper_part)

    # rank 기준으로 column-wise concat
    news_df = pd.concat(news_frames, axis=1)
    paper_df = pd.concat(paper_frames, axis=1)

    return news_df, paper_df
```

```python
years = [1991, 1992, 1993, 1994, 1995]
target = "artificial_intelligence"

news_ai, paper_ai = build_yearly_table(target, years)

print("=== NEWS ===")
display(news_ai)

print("=== PAPER ===")
display(paper_ai)
```

```python
target = "applications_of_artificial_intelligence"

news_app_ai, paper_app_ai = build_yearly_table(target, years)

print("=== NEWS ===")
display(news_app_ai)

print("=== PAPER ===")
display(paper_app_ai)
```
