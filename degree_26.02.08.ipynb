```python
import pandas as pd

targets = ["automation", "productivity", "health_care", "legislation"]
years = [1991, 1992, 1993, 1994]
topk = 20

BASE_TMPL = "/root/science-society/outputs/start=per_node__scheme=degree/mask_pred/dist=1/year={year}"

def build_rank_table(domain: str, target: str):
    # domain: "news" or "paper"
    neighbor_col = f"{domain}_neighbor"
    prob_col = f"{domain}_avg_prob"

    dfs = []
    for y in years:
        path = f"{BASE_TMPL.format(year=y)}/target={target}.tsv"
        df = pd.read_csv(path, sep="\t")

        # 컬럼 체크
        need = {"rank", neighbor_col, prob_col}
        missing = need - set(df.columns)
        if missing:
            raise ValueError(f"Missing columns {missing} in {path}. cols={list(df.columns)}")

        df = df.sort_values("rank").head(topk)

        df = df[["rank", neighbor_col, prob_col]].rename(columns={
            neighbor_col: f"{y}_token",
            prob_col: f"{y}_prob",
        }).set_index("rank")

        dfs.append(df)

    out = pd.concat(dfs, axis=1)
    out.index.name = "rank"
    return out

for target in targets:
    print(f"\n============================== TARGET: {target} ==============================")

    news_table = build_rank_table("news", target)
    paper_table = build_rank_table("paper", target)

    print("\n============================== NEWS ==============================")
    display(news_table)

    print("\n============================== PAPER ==============================")
    display(paper_table)
```
